```python
# Ορίστε τον αριθμό κόμβων n=1x0 που θα χρησιμοποιήσετε για την κατασκευή των συνθετικών τοπολογιών.
# Όπου x αντικαταστήστε το τελευταίο ψηφίο του αριθμού μητρώου σας.

n = 150
```

```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. Επιλέξτε προσεκτικά τις παραμέτρους. 
REG = nx.watts_strogatz_graph(n,d,0)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
M = 750

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n,M) 

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25
L = 1
n = 150
# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n,R) 

# για την οπτικοποίηση, χρησιμοποιήστε draw_network_edges, draw_network_nodes. 
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήστε την get_node_attributes από το networkx.
pos = nx.get_node_attributes(RGG,'pos')
nx.draw_networkx_edges(RGG,pos)
nx.draw_networkx_nodes(RGG,pos,node_size=15)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4

# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n,d)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SF)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3

# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n,d,p)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SW)
```

### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
# για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx
graphs = [REG, RGER, RGG, SF, SW]
degrees = [REG_degree, RGER_degree, RGG_degree, SF_degree, SW_degree]
def create_w_matrix(G, a, b):
    g = nx.Graph()
    for (i,j) in G.edges():
        g.add_edge(i, j, weight = np.random.rand()*(b-a) + a)
    return g

w_matrices = [create_w_matrix(G, 1, 10) for G in graphs]
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(topology):
    strengths = []
    for i in range(len(topology.adj)):
        strength = 0
        for neigh in topology.adj[i].values():
            x = list(neigh.values())
            strength = strength + x[0]
        strengths.append(strength)
    return strengths

strengths = [strength(G) for G in w_matrices]
```

##### ii) Να αναπαρασταθεί η συγκεντρωτική κατανομή δύναμης (cumulative strength distribution) των κόμβων του δικτύου (σε κάθε τιμή δύναμης αντιστοιχίζεται το πλήθος των κόμβων με δύναμη μικρότερη ή ίση από αυτή)

```python
plt.figure()
X = np.sort(strengths[0])
Y = np.array(range(n))/float(n)
plt.title("Συγκεντρωτική κατανομή δύναμης δικτύου REG")
REG_strength_sum_dist = plt.plot(Y,X)

plt.figure()
X = np.sort(strengths[1]) 
Y = np.array(range(n))/float(n)
plt.title("Συγκεντρωτική κατανομή δύναμης δικτύου RGER")
RGER_strength_sum_dist = plt.plot(X,Y)

plt.figure()
X = np.sort(strengths[2]) 
Y = np.array(range(n))/float(n)
plt.title("Συγκεντρωτική κατανομή δύναμης δικτύου RGG")
RGG_strength_sum_dist = plt.plot(X,Y)

plt.figure()
X = np.sort(strengths[3]) 
Y = np.array(range(n))/float(n)
plt.title("Συγκεντρωτική κατανομή δύναμης δικτύου SF")
SF_strength_sum_dist = plt.plot(X,Y)

plt.figure()
X = np.sort(strengths[4]) 
Y = np.array(range(n))/float(n)
plt.title("Συγκεντρωτική κατανομή βαθμού κόμβου δικτύου SW")
SW_strength_sum_dist = plt.plot(X,Y)
```

### Δ) Υπολογισμός μέσου μήκους μονοπατιού

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του στο δίκτυο. Να παρουσιαστούν συγκεντρωτικά τα αποτελέσματα. Πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως εξηγήθηκαν στη θεωρία.

```python
# χρησιμοποήστε την all_pairs_shortest_path_length του networkx
def compute_path_lengths(G):
    paths = list(nx.all_pairs_shortest_path_length(G))
    lengths = []
    for (x,y) in paths:
        lengths.append(list(y.values()))
    return lengths

REG_path_lengths = compute_path_lengths(REG)
RGER_path_lengths = compute_path_lengths(RGER)
RGG_path_lengths = compute_path_lengths(RGG)
SF_path_lengths = compute_path_lengths(SF)
SW_path_lengths = compute_path_lengths(SW)

print("Mean path length REG: %f" %np.mean(REG_path_lengths))
print("Variance of path length REG: %f" %np.var(REG_path_lengths))

print("Mean path length RGER: %f" %np.mean(RGER_path_lengths))
print("Variance of path length RGER: %f" %np.var(RGER_path_lengths))

print("Mean path length RGG: %f" %np.mean(RGG_path_lengths))
print("Variance of path length RGG: %f" %np.var(RGG_path_lengths))

print("Mean path length SF: %f" %np.mean(SF_path_lengths))
print("Variance of path length SF: %f" %np.var(SF_path_lengths))

print("Mean path length SW: %f" %np.mean(SW_path_lengths))
print("Variance of path length SW: %f" %np.var(SW_path_lengths))
```

Συγκρίνοντας τις μέσες τιμές μήκους μονοπατιού, βλέπουμε ότι το REG έχει το μεγαλύτερο μέσο μήκος μονοπατιού ($19$), αμέσως μετά είναι το SW με $4$ και μετά τα υπόλοιπα (RGER, RGG, SF), γύρω στο $2.5$.
To REG προφανως έχει τη μεγαλύτερητιμή διότι για να φτάσεις από τον έναν κόμβο στον άλλο πρέπει να διασχίσεις αναγκαστικά όλους τους κόμβους. Το SW έχει τιμή $4$ γιατί πάντα θα υπάρχουν ενδιάμεσα shortcuts-κόμβοι, που θα συντομεύουν τη διαδρομή από τον έναν κόμβο στον άλλον. Στο RGG υπάρχουν πολλοί κόμβοι που υπάρχουν κοντά μεταξύ τους, και έχουν και σχετικά μεγάλο variance ($1.4$). Τo SF έχει πάρα πολλές ακμές από τη μία πλευρά στην άλλη, οπότε λογικό είναι να έχει τόσο χαμηλή τιμή μέσου μήκους. Το RGER έχει τις περισσότερες ενδιάμεσες ακμές ($750$) όποτε είναι προφανής ο λόγος που η μέση τιμή μήκους μονοπατιού γίνεται $2.4$. 

### Ε) Υπολογισμός συντελεστή ομαδοποίησης (ΣΟ)

##### 1) Αναλυτικός υπολογισμός του ΣΟ

<h4>Αναλυτικός υπολογισμός κεντρικοτήτων</h4>
<p>Τοπολογία Α </p>
<b>Degree Centralities</b><br>
$C_1 = 3, C_2 = 1, C_3 = 1, C_4 = 1$<br>
<b>Closeness Centralities</b><br>
$C_1 = \frac{3}{1+1+1}=1, C_2 = \frac{3}{2+2+1}=\frac{3}{5}, C_3 = \frac{3}{2+2+1}=\frac{3}{5}, C_4 = \frac{3}{2+2+1}=\frac{3}{5}$<br>
<b>Betweenness Centralities</b><br>
$C_1 = \frac{1}{1}+\frac{1}{1}+ \frac{1}{1}= 3, C_2 = \frac{0}{1}+\frac{0}{1}+ \frac{0}{1}= 0, C_3 = \frac{0}{1}+\frac{0}{1}+ \frac{0}{1}= 0, C_4 = \frac{0}{1}+\frac{0}{1}+ \frac{0}{1}= 0$<br>
<p>Τοπολογία B </p>
<b>Degree Centralities</b><br>
$C_1 = 3, C_2 = 3, C_3 = 3, C_4 = 3$<br>
<b>Closeness Centralities</b><br>
$C_1 = \frac{3}{1+1+1}= 1, C_2 = \frac{3}{1+1+1}= 1, C_3 = \frac{3}{1+1+1}= 1, C_4 = \frac{3}{1+1+1}= 1$<br>

Στην τοπολογία Β, το closeness των κόμβων είναι όλων το ίδιο ($1$), διότι όλοι οι κόμβοι απέχουν το ίδιο μεταξύ τους και έχεουν και ίδιο betweenness ($0$), λογικό εφόσον ο γράφος είναι συμμετρικός.

##### 2) Υπολογισμός κεντρικότητας σε μεγαλύτερες συνθετικές τοπολογίες

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί η τοπική κεντρικότητα του κάθε κόμβου, να απεικονιστεί η συγκεντρωτική κατανομή κεντρικότητας για όλους τους κόμβους και να σχολιαστούν τα αποτελέσματα για όλες τις παρακάτω μετρικές κεντρικότητας:

- degree centrality
- closeness centrality
- betweenness centrality
- eigenvector centrality

```python
n = 150
def calculate_degree_centrality(G):
    degree_centrality = nx.degree_centrality(G)
    return degree_centrality

def calculate_closeness_centrality(G):
    closeness_centrality = nx.closeness_centrality(G)
    return closeness_centrality

def calculate_betweenness_centrality(G):
    betweenness_centrality = nx.betweenness_centrality(G)
    return betweenness_centrality

def calculate_eigenvector_centrality(G):
    eigenvector_centrality = nx.eigenvector_centrality(G,max_iter=1000)
    return eigenvector_centrality

REG_degree_centrality = calculate_degree_centrality(REG)
REG_closeness_centrality = calculate_closeness_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)
REG_eigenvector_centrality = calculate_eigenvector_centrality(REG)

RGER_degree_centrality = calculate_degree_centrality(RGER)
RGER_closeness_centrality = calculate_closeness_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)
RGER_eigenvector_centrality = calculate_eigenvector_centrality(RGER)

RGG_degree_centrality = calculate_degree_centrality(RGG)
RGG_closeness_centrality = calculate_closeness_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)
RGG_eigenvector_centrality = calculate_eigenvector_centrality(RGG)

SF_degree_centrality = calculate_degree_centrality(SF)
SF_closeness_centrality = calculate_closeness_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)
SF_eigenvector_centrality = calculate_eigenvector_centrality(SF)

SW_degree_centrality = calculate_degree_centrality(SW)
SW_closeness_centrality = calculate_closeness_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)
SW_eigenvector_centrality = calculate_eigenvector_centrality(SW)

# για πιο εύκολο σχολιασμό, να παράξετε 1 figure για όλες τις τοπολογίες ανά κεντρικότητα
def plot_centrality(centrality):
    X = np.sort(list(centrality.values()))  
    Y = np.array(range(n))/float(n)
    plt.plot(X,Y)

plt.figure()
plt.title("Degree Centrality")
plot_centrality(REG_degree_centrality)
plot_centrality(RGER_degree_centrality)
plot_centrality(RGG_degree_centrality)
plot_centrality(SF_degree_centrality)
plot_centrality(SW_degree_centrality)
plt.legend(['REG','RGER','RGG','SF','SW'])

plt.figure()
plt.title("Closeness Centrality")
plot_centrality(REG_closeness_centrality)
plot_centrality(RGER_closeness_centrality)
plot_centrality(RGG_closeness_centrality)
plot_centrality(SF_closeness_centrality)
plot_centrality(SW_closeness_centrality)
plt.legend(['REG','RGER','RGG','SF','SW'])

plt.figure()
plt.title("Betweenness Centrality")
plot_centrality(REG_betweenness_centrality)
plot_centrality(RGER_betweenness_centrality)
plot_centrality(RGG_betweenness_centrality)
plot_centrality(SF_betweenness_centrality)
plot_centrality(SW_betweenness_centrality)
plt.legend(['REG','RGER','RGG','SF','SW'])

plt.figure()
plt.title("Eigenvector Centrality")
plot_centrality(REG_eigenvector_centrality)
plot_centrality(RGER_eigenvector_centrality)
plot_centrality(RGG_eigenvector_centrality)
plot_centrality(SF_eigenvector_centrality)
plot_centrality(SW_eigenvector_centrality)
plt.legend(['REG','RGER','RGG','SF','SW'])
```

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστείτε τις adjacency_matrix (networkx), to.dense, ones (numpy)
#linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.

n = 150
d = 4
p = 0.3
M = 750
R = 0.25

REG = nx.watts_strogatz_graph(n,d,0)
RGER = nx.gnm_random_graph(n,M)
RG = nx.erdos_renyi_graph(n,p)
RGG = nx.random_geometric_graph(n,R) 
SF = nx.barabasi_albert_graph(n,d)
SW = nx.watts_strogatz_graph(n,d,p)

def calculate_ego(G):
    ego_centralities = []
    for i in range (G.number_of_nodes()):
        ego = nx.ego_graph(G,i)
        adj_ego = nx.adjacency_matrix(ego)
        adj_ego = adj_ego.todense()
        adj_ego = np.multiply(np.linalg.matrix_power(adj_ego,2),np.ones(adj_ego.shape)-adj_ego)
        adj_ego = np.triu(adj_ego,1)
        adj_ego = np.reciprocal(adj_ego, where = (adj_ego != 0))[np.nonzero(adj_ego)].sum()
        ego_centralities.append(adj_ego)
    return ego_centralities

def create_dataframe(a,b):
    dictionary = dict(zip(["Ego Betweenness Centrality","Betweenness Centrality"], [a,b]))
    dataframe = pd.DataFrame(data = dictionary)
    return dataframe

REG_ego_bet = calculate_ego(REG)
REG_bet = list(nx.betweenness_centrality(REG).values())
data_REG = create_dataframe(REG_ego_bet,REG_bet)

RGER_ego_bet = calculate_ego(RGER)
RGER_bet = list(nx.betweenness_centrality(RGER).values())
data_RGER = create_dataframe(RGER_ego_bet,RGER_bet)

RG_ego_bet = calculate_ego(RG)
RG_bet = list(nx.betweenness_centrality(RG).values())
data_RG = create_dataframe(RG_ego_bet,RG_bet)

RGG_ego_bet = calculate_ego(RGG)
RGG_bet = list(nx.betweenness_centrality(RGG).values())
data_RGG = create_dataframe(RGG_ego_bet,RGG_bet)

SF_ego_bet = calculate_ego(SF)
SF_bet = list(nx.betweenness_centrality(SF).values())
data_SF = create_dataframe(SF_ego_bet,SF_bet)

SW_ego_bet = calculate_ego(SW)
SW_bet = list(nx.betweenness_centrality(SW).values())
data_SW = create_dataframe(SW_ego_bet,SW_bet)

pairplot = sns.pairplot(data_REG).fig.suptitle("REG")
pairplot = sns.pairplot(data_RGER).fig.suptitle("RGER")
pairplot = sns.pairplot(data_RG).fig.suptitle("RG")
pairplot = sns.pairplot(data_RGG).fig.suptitle("RGG")
pairplot = sns.pairplot(data_SF).fig.suptitle("SF")
pairplot = sns.pairplot(data_SW).fig.suptitle("SW")
```
